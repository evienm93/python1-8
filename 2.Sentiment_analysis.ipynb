```python
import pandas as pd
from transformers import pipeline

df = pd.read_csv('files/detik_semua_artikel_query_clean.csv')
df.head()
```

```python
# Gunakan model sentiment Bahasa Indonesia dari HuggingFace
classifier = pipeline("sentiment-analysis", model="w11wo/indonesian-roberta-base-sentiment-classifier")
classifier
```

```python
# TESTING
kalimat = "Layanan masyarakat sangat lambat dan sop tidak jelas."
hasil = classifier(kalimat)
hasil
```

```python
# Terapkan sentiment analysis pada kolom teks
df['sentimen'] = df['judul'].apply(lambda x: classifier(str(x)[:512])[0]['label'])  # batas input 512 token
df.head()
```
